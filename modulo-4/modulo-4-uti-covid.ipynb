### Refazendo do começo, com outra visão

```python
import pandas as pd
import numpy as np
```

```python
import os
for dirname, _, filenames in os.walk('data'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        filepath = os.path.join(dirname, filename)
        break
```

```python
dados = pd.read_excel(filepath)
```

```python
dados.head()
```

```python
features_continuas = dados.iloc[:, 13:-2].columns # apenas colunas não discretas
features_continuas
```

```python
def preenche_tabela(dados):
    features_continuas_colunas = dados.iloc[:, 13:-2].columns
    features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER', as_index=False)[features_continuas_colunas].fillna(method='ffill').fillna(method='bfill')
    
    features_categoricas = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]
    dados_finais = pd.concat([features_categoricas, features_continuas, saida], 
                             ignore_index = True,
                             axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
```

```python
dados_limpos = preenche_tabela(dados)
dados_limpos.head()
```

```python
dados_limpos.describe()
```

```python
dados_limpos = dados_limpos.dropna()
```

Não podemos usar quem já entrou direto na UTI.

```python
a_remover = dados_limpos.query("WINDOW=='0-2' and ICU==1")['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos = dados_limpos.query('PATIENT_VISIT_IDENTIFIER not in @a_remover')
```

```python
dados_limpos.describe()
```

Agora precisamos definir das 5 janelas, quais eu vou usar.

```python
# feature: usaremos apenas a primeira janela, com uma feature binária caso tenha ido em qualquer janela na UTI.

def prepare_window(rows):
    # rows é o conjunto de 5 linhas para cada paciente
    # se qualquer linha deste paciente foi para UTI
    if (np.any(rows["ICU"])):
        rows.loc[rows["WINDOW"]=="0-2", "ICU"] = 1
    
    # redução das linhas retorna apenas 1 nova linha
    return rows.loc[rows["WINDOW"] == "0-2"]
```

Super desafio 6: não usar backfill da janela após a pessoa ter ido pra UTI

min 34:30 - inverter. Primeiro chamar o prepare window, add algo no meio e depois fazer o backfill.    

```python
dados_limpos = dados_limpos.groupby('PATIENT_VISIT_IDENTIFIER').apply(prepare_window)
dados_limpos.head()
```

Pacientes sem nulos e que não foram internados nas duas primeiras horas (primeira janela)

```python
x_columns = dados.describe().columns
# x_columns, dados_limpos.describe().columns.size
```

```python
# y = dados_limpos.ICU
# x = dados_limpos[x_columns].drop(['ICU'], axis='columns')
```

```python
y = dados_limpos["ICU"]
x = dados_limpos[x_columns].drop(["ICU"], axis=1)
```

```python
np.random.seed(73246)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y)

for i in (x_train, x_test, y_train, y_test):
    print(i.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.dummy import DummyClassifier

modelo = DummyClassifier()
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter = 1500)
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```
